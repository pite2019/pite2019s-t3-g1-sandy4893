```python
import sklearn
import sklearn.datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
df = pd.DataFrame(data=boston.data, columns=boston.feature_names)
print(df.head())
```

```python
df.hist(column=["NOX","AGE"], bins=25)
```

```python
boston_target = boston.target
x = df.loc[:, ['NOX' , "AGE"]]
y = boston.target
regr = LinearRegression()
regr.fit(x,y)
```

```python
predict = regr.predict(x)
plt.scatter(x["NOX"], y)
plt.title('Scatter of "NOX" vs "Target"')
plt.show()

plt.plot(x["NOX"], predict)
plt.title('Scatter of "NOX" & predicted values of Target')
plt.show()
```

```python
predict = regr.predict(x)
plt.scatter(x["AGE"], y)
plt.title('Scatter of "AGE" vs "Target"')
plt.show()

plt.plot(x["AGE"], predict)
plt.title('Scatter of "AGE" & predicted values of Target')
plt.show()
```
